```python
# Heatmap of Average House Prices by Living Area and Number of Bedrooms

# Prepare data
data = df.loc[:, ["living area", "number of bedrooms", "Price"]]
data_pivot = data.pivot_table(values="Price", index="number of bedrooms", columns="living area")

# Create heatmap
fig, ax = plt.subplots(figsize=(10, 7))
heatmap = ax.pcolormesh(data_pivot, cmap="viridis")

# Add colorbar
fig.colorbar(heatmap, label="Price")

# Customize plot
ax.set_xlabel("Living Area")
ax.set_ylabel("Number of Bedrooms")
ax.set_title("Heatmap of House Prices")

# Show plot
plt.show()
```

```python
# 3D Scatter Plot of House Prices by Living Area and Number of Bathrooms

# Prepare data
x_values = df["living area"]
y_values = df["Price"]
z_values = df["number of bathrooms"]

# Create 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(projection="3d")

# Plot data
ax.scatter(x_values, y_values, z_values, color="blue", alpha=0.5)

# Customize plot
ax.set_xlabel("Living Area")
ax.set_ylabel("Price")
ax.set_zlabel("Number of Bathrooms")
ax.set_title("3D Scatter Plot of House Prices")

# Show plot
plt.show()
```
